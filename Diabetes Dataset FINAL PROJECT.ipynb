### Split Data

```python
y=data.loc[:,'Diabetes'].values
data.drop(['Diabetes'],axis=1,inplace=True)
X=data.iloc[:,:].values
```

```python
from sklearn.model_selection import train_test_split
train_X,test_X,train_y,test_y=train_test_split(X,y,test_size=0.2) #train 80% ,test 20%
```

## Model 
Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(train_X,train_y)#for train 
```

```python
y_pred=model.predict(test_X)
```

evaluate Model

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print(accuracy_score(y_pred,test_y))
```

```python
confusion_matrix(y_pred,test_y)
```

### SVM 

```python
from sklearn.svm import SVC
model2=SVC(kernel='rbf')
```

```python
model2.fit(train_X,train_y)
```

```python
y_pred2=model2.predict(test_X)
print("Accuracy: ",accuracy_score(y_pred2,test_y)*100,' %')
```
